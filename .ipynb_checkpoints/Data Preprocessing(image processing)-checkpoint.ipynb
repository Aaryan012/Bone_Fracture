```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r'Image_labels_path.csv')
```

```python
data
data.rename(columns={'combine_fracture': 'fracture'}, inplace=True)
```

```python
data['fracture']
```

```python
def image_processing(img_path):
    img=cv2.imread(img_path)
    plt.imshow(img)
    # convert to rgb
    img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    # resize to 100*100
    img_res=cv2.resize(img_rgb,(100,100))
    plt.imshow(img_res)
    # convert to numpy array
    img_ary=np.array(img_res).flatten()
    # convert to dataFrame
    img_df=pd.DataFrame(img_ary).T
    return img_df
```

```python
image_processing(data['fracture'][0])
```

```python
final_df=pd.DataFrame()
```

```python
for img in data['fracture']:
    df_x=image_processing(img)
    final_df=pd.concat([final_df,df_x])
final_df['label']='Fractured'
```

```python
final_df.index=range(len(final_df))
```

```python
final_df.to_csv('bone_fracture_unlabelled.csv',index=False)
```
